```python
import pandas as pd
import numpy as np
import sklearn as sk
import math 

import nltk
# # nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

```python
#I'm using a dictionary corpus
corpus = {"Doc1": "Information Retrieval System",
        "Doc2": "Information about the Human Resource System",
        "Doc3": "Resources allocated to Faculty of Computing Information Centre",
        "Doc4": "Ranked Retrieval Model with High Accuracy",
        "Doc5": "Boolean Retrieval Model has Several Problems"}


# TO USE A LIST CORPUS
# corpus = ["I want to start learning to charge something in life",
# "reading something about life no one else knows",
#  "Never stop learning"]
# corpusDict = {}  
# for i,k in enumerate(corpus):
#         corpusDict["doc"+str(i)] = k
# corpus = corpusDict 
```

```python
def TF(corpus):

    total = set(np.concatenate([ i.split(' ') for  i in  corpus.values()]).flat)

    TC = {}
    for i in corpus.keys():
        wordDict = dict.fromkeys(total, 0)
        for word in corpus[i].split(' '):
            wordDict[word]+=1
        TC[i] = wordDict
        

    logTF = {}
    for i in corpus.keys():
        wordDict = dict.fromkeys(total, 0)
        for word, count in TC[i].items():
            if count > 0:
                wordDict[word] = 1 + math.log10(count)
        logTF[i] = wordDict

    # TF = {}
    # for i in corpus.keys():
    #     wordDict = dict.fromkeys(total, 0)
    #     for word, count in TC[i].items():
    #         if count > 0:
    #             wordDict[word] = count/len(corpus[i].split(' '))
    #     TF[i] = wordDict
    # return(TC,TF,total)

    return(TC,logTF, total)


TC,logTF,total = TF(corpus)
pd.DataFrame([i for i in logTF.values()], index = logTF.keys())
pd.DataFrame([i for i in logTF.values()], index = logTF.keys())
```

```python
def computeIDF(docList):
    idfDict = dict.fromkeys(docList[0].keys(), 0)
    
    for doc in docList:
        for word, val in doc.items():
            if val > 0:
                idfDict[word] += 1
    print(idfDict)
    for word, val in idfDict.items():
        idfDict[word] = 1+ math.log10( len(docList) / (float(val) ))
        
    return(idfDict)

#inputing our sentences in the log file
idfs = computeIDF(list(TC.values()))
pd.DataFrame([idfs])
```

```python
def computeTFIDF(logTF, idfs):
    tfidf = {}
    for word, val in logTF.items():
        
        tfidf[word] = val*idfs[word]
        
    return(tfidf)

tdidf = pd.DataFrame([computeTFIDF(logTF[i], idfs) for i in logTF.keys()], index = logTF.keys())
tdidf
```

### Normalizing

```python
normalized_corpus = corpus.copy()
for key,value in corpus.items():
    normalized_corpus[key] = value.lower()

# UNCOMMENT TO REMOVE STOP WORDS
# for key, value in normalized_corpus.items():
#     normalized_corpus[key] = value.lower().split()
# for key, value in normalized_corpus.items():
#         normalized_corpus[key] = " ".join([word for word in value if word not in stop_words])
# print(normalized_corpus)


normalized_logTF, norm_total = TF(normalized_corpus)[1:]
normalized_idfs = computeIDF(list(normalized_logTF.values()))
normalized_tdidf = pd.DataFrame([computeTFIDF(normalized_logTF[i], normalized_idfs) for i in normalized_logTF.keys()], index = normalized_logTF.keys())

#length normalization of the tfidf of documents
normalized_length_tdidf =  normalized_tdidf.div(normalized_tdidf.pow(2).sum(axis = 1).pow(0.5), axis = 0)
print(normalized_length_tdidf)
```

### Query Tf-Idf

```python
query = "Information Retrieval"
query = query.lower()

# UNCOMMENT TO REMOVE STOP WORDS FOR QUERY
# remove stop words from query
# query = query.lower().split()
# query = " ".join([word for word in query if word not in stop_words])

queryWordsInDocs = [i for i in norm_total if i in query.lower().split(' ')]
queryWordsNOtInDocs = [i for i in total if i in query.lower().split(' ')]

query_df = pd.DataFrame([i for i in normalized_logTF.values()], index = normalized_logTF.keys())[queryWordsInDocs]
query_norm_tf = TF({"query": query.lower()})[1]["query"]
query_norm_tfidf = pd.DataFrame(computeTFIDF(query_norm_tf, normalized_idfs),index = [0])

# query_tfidf =  query_tfidf[queryWordsInDocs]
query_tfidf =  normalized_length_tdidf[queryWordsInDocs+queryWordsNOtInDocs]

print(query_df,'\n',query_tfidf)
print("\nTF*IDF for the query")
query_norm_tfidf
```

### Cosine Similarity & Ranking

```python
# Cosine for length-normalized vectors

def cosine_similarity(vec1, vec2):
    # normalize to unit length
    norm_vec1 = vec1 / math.sqrt((vec1**2).sum(axis=1))
    norm_vec2 = vec2 / math.sqrt((vec2**2).sum())
    # compute dot product
    dot_product = np.dot(vec1, vec2)
    return dot_product


query_tfidf_cosine= query_tfidf.copy()
query_tfidf_cosine['doc'] = corpus.values()

#Cosine similarity
query_tfidf_cosine["Cosine_distance"] = [cosine_similarity(query_norm_tfidf, query_tfidf.iloc[i]) for i in range(len(query_tfidf) ) ]

query_tfidf_cosine.sort_values(by = "Cosine_distance", ascending = False)
```
